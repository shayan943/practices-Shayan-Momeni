```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,10,100)
y = np.sin(x)
plt.figure(figsize=(6,4))
plt.plot(x,y,label='sin(x)',color='blue')
plt.title('plot of sin(x) - Low Resolution')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(0,10,100)
y = 2 * np.sin(x/2) * (np.cos(x/4)**2)

plt.figure(figsize=(6,4))
plt.plot(x,y,label='2*sin(x/2)*cos(x/4)**2-Red' , color = 'red')
plt.title('plot of 2*sin(x/2)*cos(x/4)**2 - Low Resolution')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = 2 * np.sin(x/2) * (np.cos(x/4)**2)

plt.figure(figsize=(6,4))
plt.plot(x,y1,label='sin(x)-Blue' , color='blue')
plt.plot(x,y2,label='2*sin(x/2)*cos(x/4)**2-Red', color = 'red')
plt.title('comparison of two functions')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
```
