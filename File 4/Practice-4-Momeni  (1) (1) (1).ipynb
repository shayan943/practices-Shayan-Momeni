```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# خواندن تصویر 
img_path = "dl2.jpg"
img = cv2.imread(img_path)
# تصویر اصلی
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# تصویر خاکستری
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# تصویر معکوس
inverted_img = cv2.bitwise_not(gray)
# thershold تصویر
ret, threshold_img = cv2.threshold(inverted_img, 128,255, cv2.THRESH_BINARY)
# تصویر وارونه 
thresh_flip_vertical = cv2.flip(threshold_img, 0)
# چهار ردیف و دو ستون برای تصاویر
fig, axs = plt.subplots(4,2, figsize=(12,16))
fig.suptitle("Image Processing", fontsize=18)
# ردیف 1
axs[0,0].imshow(img_rgb)
axs[0,0].set_title("Original Image")
axs[0,0].axis("off")
axs[0,1].axis("off")

#axs[0,1].hist(gray.ravel(), bins=256 , color='black')
#axs[0,1].set_title("Histogram of Gray")
#axs[0,1].axis("off")
# ردیف 2
axs[1,0].imshow(gray, cmap="gray")
axs[1,0].set_title("Gray Image")
axs[1,0].axis("off")
axs[1,1].hist(gray.ravel(), bins=256, color='black')
axs[1,1].set_title("Gray Histogram")
#axs[1,1].axis("off") 
# ردیف 3
axs[2,0].imshow(inverted_img, cmap="gray")
axs[2,0].set_title("Inverted Image")
axs[2,0].axis("off")
axs[2,1].hist(inverted_img.ravel(), bins=256, color="black")
axs[2,1].set_title(" Inverted Histogram")
# ردیف 4
axs[3,0].imshow(thresh_flip_vertical, cmap='gray')
axs[3,0].set_title("Threshold Image")
axs[3,0].axis("off")
axs[3,1].hist(threshold_img.ravel(), bins=256, color='black')
axs[3,1].set_title("Threshold Histogram")
# تنظیم فاصله ها 
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
