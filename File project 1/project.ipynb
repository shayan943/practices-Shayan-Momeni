```python
import cv2
import matplotlib.pyplot as plt


class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.borrowed = False

    def show_info(self):
        status = "امانت داده شده" if self.borrowed else "موجود"

        print("------------------")
        print("شناسه:", self.book_id)
        print("نام کتاب:", self.title)
        print("نویسنده:", self.author)
        print("وضعیت:", status)


class Library:
    def __init__(self):
        self.books = []

    #  اضافه کردن کتاب 
    def add_book(self):
        book_id = input("شناسه کتاب: ")

        # جلوگیری از تکرار ID
        for book in self.books:
            if book.book_id == book_id:
                print("این شناسه قبلاً ثبت شده است.")
                return

        title = input("نام کتاب: ")
        author = input("نام نویسنده: ")

        self.books.append(Book(book_id, title, author))
        print("کتاب اضافه شد.")

    #  نمایش کتاب‌ها 
    def show_books(self):
        if not self.books:
            print("کتابی وجود ندارد.")
            return

        for book in self.books:
            book.show_info()

    #  جستجو 
    def search_book(self):
        title = input("نام کتاب: ")

        found = False

        for book in self.books:
            if title.lower() in book.title.lower():
                book.show_info()
                found = True

        if not found:
            print("کتاب پیدا نشد.")

    #  حذف کتاب 
    def remove_book(self):
        book_id = input("شناسه کتاب: ")

        for book in self.books:
            if book.book_id == book_id:
                self.books.remove(book)
                print("کتاب حذف شد.")
                return

        print("کتاب پیدا نشد.")

    #  امانت گرفتن 
    def borrow_book(self):
        book_id = input("شناسه کتاب: ")

        for book in self.books:
            if book.book_id == book_id:

                if book.borrowed:
                    print("این کتاب قبلاً امانت گرفته شده است.")
                else:
                    book.borrowed = True
                    print("کتاب با موفقیت امانت داده شد.")

                return

        print("کتاب پیدا نشد.")

    #  بازگرداندن 
    def return_book(self):
        book_id = input("شناسه کتاب: ")

        for book in self.books:
            if book.book_id == book_id:

                if book.borrowed:
                    book.borrowed = False
                    print("کتاب بازگردانده شد.")
                else:
                    print("این کتاب امانت داده نشده است.")

                return

        print("کتاب پیدا نشد.")


   #بخش خواندن تصویر  cv2 + matplotlib
    

    def image_analysis(self):
        path = input("مسیر تصویر کتاب: ")

        img = cv2.imread(path)

        if img is None:
            print("تصویر پیدا نشد!")
            return

        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        blur = cv2.GaussianBlur(gray, (5, 5), 0)

        thresh = cv2.adaptiveThreshold(
            blur,
            255,
            cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
            cv2.THRESH_BINARY,
            11,
            2
        )

        # هیستوگرام 
        hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

        plt.figure()

        plt.title("Histogram")
        plt.plot(hist)
        plt.show()

        #  نمایش تصاویر 
        plt.figure(figsize=(16,10))

        plt.subplot(1, 3, 1)
        plt.title("Original")
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

        plt.subplot(1, 3, 2)
        plt.title("Gray")
        plt.imshow(gray, cmap='gray')

        plt.subplot(1, 3, 3)
        plt.title("Threshold")
        plt.imshow(thresh,cmap='gray')

        plt.show()


# اجرا 

library = Library()

while True:
    print("\n===== سیستم مدیریت کتابخانه =====")
    print("1. افزودن کتاب")
    print("2. نمایش کتاب‌ها")
    print("3. جستجوی کتاب")
    print("4. حذف کتاب")
    print("5. امانت گرفتن کتاب")
    print("6. بازگرداندن کتاب")
    print("7. تحلیل تصویر کتاب (cv2)")
    print("8. خروج")

    choice = input("انتخاب شما: ")

    if choice == "1":
        library.add_book()

    elif choice == "2":
        library.show_books()

    elif choice == "3":
        library.search_book()

    elif choice == "4":
        library.remove_book()

    elif choice == "5":
        library.borrow_book()

    elif choice == "6":
        library.return_book()

    elif choice == "7":
        library.image_analysis()

    elif choice == "8":
        print("خروج از برنامه")
        break

    else:
        print("گزینه نامعتبر است.")
```
